```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input, decode_predictions
```

```python
im = image.load_img(r'laptop.png', target_size=(224, 224))
# also try: monkey.jpg , bottle.png, laptop.png
im
```

```python
model = MobileNetV2(include_top=True)
```

```python
a = image.img_to_array(im)  # im is an image object -> numpy array
print(a.shape)
```

```python
#a = a[::-1, :, :]
#from PIL import Image
#Image.fromarray(a, mode="RGB")
```

```python
print(a.min(), a.max())
```

```python
X = preprocess_input(a)
X = X.reshape(1, 224, 224, 3)
print(X.shape)
```

```python
print(X.min(), X.max())
```

```python
result = model.predict(X)
```

```python
[f"{category:25}  {prob:6.3f}" for _, category, prob in decode_predictions(result)[0]]
```
