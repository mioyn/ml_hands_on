```python
# Step 1: Install required packages
!pip install opencv-python-headless matplotlib ipywidgets

import cv2
import matplotlib.pyplot as plt
from google.colab import files
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output

# Step 2: Upload an image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
image = cv2.imdecode(np.frombuffer(uploaded[image_path], np.uint8), cv2.IMREAD_COLOR)

# Step 3: Define OpenCV effects
def apply_effect(effect_name):
    img = image.copy()
    output_img = None

    if effect_name == 'Original':
        output_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    elif effect_name == 'Grayscale':
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        output_img = gray

    elif effect_name == 'Resized':
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        output_img = cv2.resize(gray, (300, 300))

    elif effect_name == 'Rectangle':
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        rect_img = cv2.cvtColor(cv2.resize(gray, (300, 300)), cv2.COLOR_GRAY2BGR)
        cv2.rectangle(rect_img, (50, 50), (200, 200), (0, 255, 0), 3)
        output_img = rect_img

    elif effect_name == 'Edges':
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        edges = cv2.Canny(gray, 100, 200)
        output_img = edges

    elif effect_name == 'Blurred':
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        blurred = cv2.GaussianBlur(gray, (7, 7), 0)
        output_img = blurred

    elif effect_name == 'Threshold':
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        _, thresh = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY)
        output_img = thresh

    elif effect_name == 'Text':
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        text_img = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)
        cv2.putText(text_img, 'OpenCV Demo', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
        output_img = text_img

    elif effect_name == 'Color Filter':
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        color_filter = cv2.applyColorMap(gray, cv2.COLORMAP_JET)
        output_img = color_filter

    # Display the output image
    plt.figure(figsize=(6,6))
    if len(output_img.shape) == 2:
        plt.imshow(output_img, cmap='gray')
    else:
        plt.imshow(cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title(effect_name)
    plt.show()

# Step 4: Create dropdown for selecting effects
effect_dropdown = widgets.Dropdown(
    options=['Original', 'Grayscale', 'Resized', 'Rectangle', 'Edges', 'Blurred', 'Threshold', 'Text', 'Color Filter'],
    description='Effect:'
)

# Step 5: Link dropdown to function
out = widgets.Output()

def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        with out:
            clear_output(wait=True)
            apply_effect(change['new'])

effect_dropdown.observe(on_change)
display(effect_dropdown, out)

# Step 6: Initialize with original image
with out:
    apply_effect('Original')
```
